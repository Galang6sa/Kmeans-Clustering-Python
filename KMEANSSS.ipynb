```python
#mengatasi outlier

#menggantikan nilai outlier dengan nilai threshold
def outlier_threshold(df, col, q1=0.25, q3=0.75):
    quartile1 = df[col].quantile(q1) #25% batas bawah
    quartile3 = df[col].quantile(q3) #75% batas atah
    interquartil_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquartil_range #rumus
    low_limit = quartile1 - 1.5 * interquartil_range #rumus
    return low_limit, up_limit

def replace_with_threshold(df, col):
    low, up = outlier_threshold(df, col)
    df[col] = np.where(df[col] > up, up , np.where(df[col] < low, low, df[col]))
    return df

for i in df.columns:
    df = replace_with_threshold(df, i)
```

```python
#membuat centroid acak
def random_centroid(df, k):
    centroid = []
    for _ in range(k):
        centroid.append(df.sample().values[0])
    return centroid
```

```python
#Menentukan cluster berdasarkan jarak euclidean
def assign_cluster(df, initial_centroid, k):
    cluster = []
    for i in range(len(df)):
        euclidean_distance = []
        for j in range(k):
            # Menghitung jarak Euclidean antara titik data i dan centroid j
            euclidean_distance.append(np.linalg.norm(df.iloc[i] - initial_centroid[j]))
        # Memilih indeks centroid dengan jarak terdekat sebagai cluster untuk data i
        cluster.append(np.argmin(euclidean_distance))
    return cluster
```

```python
#menghitung centroid baru

def compute_centroid(df, cluster, k):
    centroids = []
    for i in range(k):
        # Mengumpulkan semua data yang termasuk dalam cluster ke-i
        temp = []
        for j in range(len(df)):
            if cluster[j] == i:
                temp.append(df.iloc[j])
        
        # Menghitung rata-rata dari semua data dalam cluster ke-i
        if temp:  # Pastikan cluster tidak kosong
            centroids.append(np.mean(temp, axis=0))
        else:
            # Jika cluster kosong, centroid tidak berubah atau tetap
            centroids.append(np.zeros(df.shape[1]))  # atau metode lainnya untuk menangani cluster kosong
            
    return centroids
```

```python
#mengecek perbedaan antara centroid lama dan baru

def difference(old, new):
    diff = 0
    for i in range(len(old)):
        # Menghitung jarak Euclidean antara centroid lama dan baru
        diff += np.linalg.norm(np.subtract(old[i], new[i]))
    return diff
```

```python
def kmeans(df, k):
    # Inisialisasi centroid secara acak
    initial_centroid = random_centroid(df, k)
    
    # Mengelompokkan data berdasarkan centroid awal
    cluster = assign_cluster(df, initial_centroid, k)
    
    # Menghitung centroid baru berdasarkan cluster awal
    new_centroid = compute_centroid(df, cluster, k)
    
    # Loop hingga perbedaan antara centroid lama dan baru lebih kecil dari 0.001
    while difference(initial_centroid, new_centroid) > 0.001:
        # Perbarui centroid lama dengan centroid baru
        initial_centroid = new_centroid
        
        # Mengelompokkan kembali data berdasarkan centroid yang diperbarui
        cluster = assign_cluster(df, initial_centroid, k)
        
        # Hitung centroid baru lagi berdasarkan cluster yang diperbarui
        new_centroid = compute_centroid(df, cluster, k)
    
    # Menambahkan hasil cluster ke kolom baru dalam DataFrame
    df['cluster'] = cluster
    
    return df
```

```python
#menghitung jumlah lcuster
def inertia(df, cluster, k):
    wcss = []
    for i in range(k):
        temp = []
        for j in range(len(df)):
            if cluster[j] == i:
                temp.append(df.iloc[j])
        wcss.append(np.sum(np.linalg.norm(np.subtract(temp, np.mean(temp, axis=0)), axis=1)))
    return np.sum(wcss)
```
